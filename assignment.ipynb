# PLOTTING VARIOUS GRAPHS TO ANALYSE GIVEN DATA


```python
sns.histplot(df['age'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='sex', data=df,hue='slope')
plt.title('Count plot of male and female')
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.boxplot(x='sex', y='cholestoral', data=df)
plt.title('sex VS cholestoral')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='chest_pain_type', y='Max_heart_rate', data=df)
plt.title('Max heart rate vs type of chest pain')
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.countplot(x='thalassemia', hue='target', data=df)
plt.title("Distribution of Thalassemia by Target")
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.countplot(x='fasting_blood_sugar',hue='chest_pain_type', data=df)
plt.title("Distribution of Fasting blood sugar by Type of chest pain")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='exercise_induced_angina', y='oldpeak', data=df)
plt.title("Exercise_induced_angina VS Oldpeak")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='vessels_colored_by_flourosopy',data=df,hue='exercise_induced_angina')

plt.title("Distribution of Fluoroscopy-Visible Vessels by Exercise-Induced Angina")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='rest_ecg',hue='slope',data=df)
plt.title('Distribution of rest_ecg by slope')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='resting_blood_pressure', y='Max_heart_rate', data=df)
plt.title('Resting_blood_pressure VS Max_heart_rate')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='sex',y='vessels_colored_by_flourosopy',data=df)
plt.title('Distribution of vessels_colored_by_flourosopy VS sex')
plt.show()
```
